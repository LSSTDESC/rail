## Print "Welcome" statement in a for loop

Now we want to print out statements in a for loop, substituting in a variable so we know which iteration printed out the statement. We will use python's "f-strings" formatting for this (see [this primer](https://fstring.help/) for more details about how those work):

```python
for i in range(0,3): 
    print(f"Welcome to world {i}")
```

Now we have a print statement for each of the 3 iterations of the loop, starting with 0 because our loop iteration started at 0.

## Add two arrays together

Now we're going to use `numpy` to add two arrays together (see [`numpy` docs](https://numpy.org/doc/stable/reference/generated/numpy.add.html#numpy.add) for more details). First we need to import numpy, and create our two arrays:

```python
import numpy as np

arr_1 = np.random.randn(4)
arr_2 = np.random.randn(4)

print(f"array 1: {arr_1}")
print(f"array 2: {arr_2}")
```

We've got two arrays now of the same length, so if we add these together we'll just get an array of the same length with each value being the sum of the corresponding values in the two arrays we're adding:

```python
added_arr = arr_1 + arr_2
print(added_arr)
```

Now let's plot the sum of the two arrays versus the first array to get a sense of what this looks like: 

```python
import matplotlib.pyplot as plt 

plt.scatter(arr_1, added_arr)
plt.xlabel("first array")
plt.ylabel("sum of arrays")
```

Now we can see how the values of the first array compare to the sum of both arrays.
